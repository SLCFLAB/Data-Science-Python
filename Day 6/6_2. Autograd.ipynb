# 2. Autograd

```python
import torch
```

## 2.1 Autograd 

```python
x = torch.ones(1)
x.requires_grad
```

```python
y = torch.ones(1)
y.requires_grad
```

```python
z = x + y
```

```python
z.requires_grad
```

```python
x.requires_grad=True
# x.requires_grad_()
x.requires_grad
```

```python
z = x + y
```

```python
z.requires_grad
```

## 2.2 Backward

```python
y = torch.tensor([[1, 2], [3, 4]], dtype=torch.float32, requires_grad=True)
y
```

```python
y.sum()
```

```python
y.max()
```

```python
z = 2*y + 1
z
```

```python
print("y.requires_grad :", y.requires_grad)
print("z.requires_grad :", z.requires_grad)

print("")

print("y.grad :", y.grad)
print("z.grad :", z.grad)

print("")

print("y.grad_fn :", y.grad_fn)
print("z.grad_fn :", z.grad_fn)
```

```python
gz = z.grad_fn
print("z.grad_fn :", gz)
print("Where is z.grad_fn?", id(gz))
print("Where is z.grad_fn in hex?", hex(id(gz)))
```

```python
# grad can be implicitly created only for scalar outputs
z.backward()
```

```python
out = z.sum()
out
```

```python
print("y.requires_grad :", y.requires_grad)
print("z.requires_grad :", z.requires_grad)
print("out.requires_grad :", out.requires_grad)

print("")

print("y.grad :", y.grad)
print("z.grad :", z.grad)
print("out.grad :", out.grad)

print("")

print("y.grad_fn :", y.grad_fn)
print("z.grad_fn :", z.grad_fn)
print("out.grad_fn :", out.grad_fn)
```

```python
out.backward()
```

```python
print("y.requires_grad :", y.requires_grad)
print("z.requires_grad :", z.requires_grad)
print("out.requires_grad :", out.requires_grad)

print("")

print("y.grad :", y.grad)
print("z.grad :", z.grad)
print("out.grad :", out.grad)

print("")

print("y.grad_fn :", y.grad_fn)
print("z.grad_fn :", z.grad_fn)
print("out.grad_fn :", out.grad_fn)
```

```python
# By default, gradients are only retained for leaf variables. non-leaf variables
# gradients are not retained to be inspected later. This was done by design, to save memory.
print("y == leaf? : ", y.is_leaf)
print("z == leaf? : ", z.is_leaf)
```

## 2.3 Detach from Autograd & .detach() vs .data

```python
# Graph 위에 있는 변수는 바로 numpy로 변환할 수 없음.
z.numpy()
```

```python
# 따라서 .detach와 .data 중 하나를 사용해야함.
z.detach().numpy()
```

```python
# 따라서 .detach와 .data 중 하나를 사용해야함.
z.data.numpy()
```
