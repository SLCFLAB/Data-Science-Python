## 1. Sequence Classification(ë¬¸ìì—´ ë¶„ë¥˜)

> ë¬¸ìì—´ ë¶„ë¥˜ëŠ” ì£¼ì–´ì§„ ë²”ì£¼ì˜ ê°œìˆ˜ì— ë”°ë¼ ë¬¸ìì—´ì„ ë¶„ë¥˜í•˜ëŠ” ì‘ì—…ìœ¼ë¡œ **GLUE(General Language Understanding Evaluation) ë°ì´í„°ì…‹**ì´ ê·¸ ì˜ˆë¼ê³  í•  ìˆ˜ ìˆìŒ. íŒŒì´í”„ë¼ì¸ì„ ì‚¬ìš©í•´ì„œ ê°ì • ë¶„ì„ì„ í•œ ì‚¬ë¡€ê°€ ì•„ë˜ì— ìˆìœ¼ë©°, ê°ê°ì˜ ë¬¸ì¥ì´ ê¸ì •ì˜ ì˜ë¯¸ì¸ì§€ ë¶€ì •ì˜ ì˜ë¯¸ì¸ì§€ë¥¼ íŒë³„í•¨.(ì•„ë˜ì™€ ê°™ì´ ê¸ì •/ë¶€ì •ì˜ ë¼ë²¨ì„ ì ìˆ˜ì™€ í•¨ê»˜ ì¶œë ¥í•´ëƒ„) (GLUE ë°ì´í„°ì…‹ì— ëŒ€í•œ ë” ìì„¸í•œ ì„¤ëª…ì€ https://huffon.github.io/2019/11/16/glue/ë¥¼ ì°¸ê³ !)

```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
result = classifier("I hate you")[0]
print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
result = classifier("I love you")[0]
print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```

>**ì•„ë˜ëŠ” ë‘ ë¬¸ì¥ì´ ì„œë¡œ íŒ¨ëŸ¬í”„ë ˆì´ì§•ì´ ëœ ë¬¸ì¥ì¸ì§€ë¥¼ íŒë³„í•˜ëŠ” ë¬¸ìì—´ ë¶„ë¥˜ì˜ í•œ ê°€ì§€ ì‚¬ë¡€ë¡œì„œ ê³¼ì •ì€ ë‹¤ìŒê³¼ ê°™ìŒ.** 

1. í† í¬ë‚˜ì´ì €ë¥¼ ì‘ë™í•˜ê³  ì²´í¬ í¬ì¸íŠ¸ ì´ë¦„ìœ¼ë¡œ ëª¨ë¸ì„ ì‹œì‘í•¨. ëª¨ë¸ì€ **BERT**ëª¨ë¸ì´ë©° ì²´í¬í¬ì¸íŠ¸ ë‚´ì—    ì €ì¥ë˜ì–´ ìˆëŠ” ê°€ì¤‘ì¹˜ë¡œ í•¨ê»˜ ì´ë¥¼ ì—…ë¡œë“œí•¨
2. ë‘ ë¬¸ì¥ì—ì„œë¶€í„° ëª¨ë¸ì— í•œì •ëœ êµ¬ë¶„ì, í† í° ì¢…ë¥˜, (í† í¬ë‚˜ì´ì €ì— ì˜í•´ì„œ ìë™ìœ¼ë¡œ ìƒì„±ë ) ì–´í…ì…˜   ë§ˆìŠ¤í¬ë¥¼ ê°€ì§€ê³  ë¬¸ìì—´ì„ ë§Œë“¦
3. í•´ë‹¹ ë¬¸ìì—´ì„ ëª¨ë¸ì— ì§‘ì–´ë„£ì–´ì„œ ë‘ ê°œì˜ ë²”ì£¼ (0: íŒ¨ëŸ¬í”„ë ˆì´ì§•ì„ ì•ˆ í–ˆìŒ, 1: íŒ¨ëŸ¬í”„ë ˆì´ì§•ì„ í–ˆìŒ) ì¤‘ì˜ í•˜ë‚˜ë¡œ ë¶„ë¥˜ë˜ë„ë¡ í•¨
4. ë²”ì£¼ë“¤ì— ëŒ€í•œ í™•ë¥ ê°’ì„ ì–»ê¸° ìœ„í•´ì„œ ê²°ê³¼ì˜ ì†Œí”„íŠ¸ë§¥ìŠ¤ê°’ì„ êµ¬í•¨
5. ê²°ê³¼ë¥¼ ì¶œë ¥í•¨

```python
# ëª¨ë“ˆ ë¶ˆëŸ¬ì˜¤ê¸° 
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# í† í¬ë‚˜ì´ì €ì™€ ëª¨ë¸ ì§€ì •
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased-finetuned-mrpc")
model = AutoModelForSequenceClassification.from_pretrained("bert-base-cased-finetuned-mrpc")


# ë²”ì£¼ ì„¤ì •
classes = ["not paraphrase", "is paraphrase"]

# ë¬¸ìì—´ ì„¤ì •
sequence_0 = "The company HuggingFace is based in New York City" # ë¬¸ìì—´_0: "íšŒì‚¬ í—ˆê¹…í˜ì´ìŠ¤ëŠ” ë‰´ìš• ë„ì‹œì— ìœ„ì¹˜í•´ ìˆë‹¤.""

sequence_1 = "Apples are especially bad for your health" # ë¬¸ìì—´_1 "ì‚¬ê³¼ëŠ” íŠ¹íˆ ë‹¹ì‹ ì˜ ê±´ê°•ì— ì¢‹ì§€ ì•Šë‹¤."

sequence_2 = "HuggingFace's headquarters are situated in Manhattan" # ë¬¸ìì—´_2"í—ˆê¹…í˜ì´ìŠ¤ ì‚¬ì˜ ë³¸ë¶€ëŠ” ë§¨í•˜íƒ„ì— ìœ„ì¹˜í•´ìˆë‹¤."



# í† í¬ë‚˜ì´ì €ëŠ” ìë™ìœ¼ë¡œ ì–´í…ì…˜ ë§ˆìŠ¤í¬ë¥¼ ê³„ì‚°í•  ë¿ë§Œ ì•„ë‹ˆë¼ êµ¬ë¶„ì ì¤‘ì˜ í•˜ë‚˜( <CLS>ì™€ <SEP>ì™€ í† í°ì„ ë¬¸ìì—´ì— ë„£ìŒ 

paraphrase = tokenizer(sequence_0, sequence_2, return_tensors="pt") # ë¬¸ìì—´_0ê³¼ ë¬¸ìì—´_2ê°„ì˜ íŒ¨ëŸ¬í”„ë ˆì´ì§• í™•ë¥ ì„ ê³„ì‚° 

not_paraphrase = tokenizer(sequence_0, sequence_1, return_tensors="pt") # ë¬¸ìì—´_0ê³¼ ë¬¸ìì—´_1ê°„ì˜ íŒ¨ëŸ¬í”„ë ˆì´ì§• í™•ë¥ ì„ ê³„ì‚° 



paraphrase_classification_logits = model(**paraphrase).logits # ë¬¸ìì—´_0ê³¼ ë¬¸ìì—´_2ê°„ì˜ íŒ¨ëŸ¬í”„ë ˆì´ì§• í™•ë¥ ì„ ë¡œì§“ í•¨ìˆ˜ê°’ìœ¼ë¡œ ê³„ì‚° 

not_paraphrase_classification_logits = model(**not_paraphrase).logits # ë¬¸ìì—´_0ê³¼ ë¬¸ìì—´_1ê°„ì˜ íŒ¨ëŸ¬í”„ë ˆì´ì§• í™•ë¥ ì„ ë¡œì§“ í•¨ìˆ˜ê°’ìœ¼ë¡œ ê³„ì‚° 



paraphrase_results = torch.softmax(paraphrase_classification_logits, dim=1).tolist()[0] # ë¬¸ìì—´_0ê³¼ ë¬¸ìì—´_2ê°„ì˜ íŒ¨ëŸ¬í”„ë ˆì´ì§• í™•ë¥  ê²°ê³¼ê°’ ë„ì¶œ

not_paraphrase_results = torch.softmax(not_paraphrase_classification_logits, dim=1).tolist()[0] # ë¬¸ìì—´_0ê³¼ ë¬¸ìì—´_1ê°„ì˜ íŒ¨ëŸ¬í”„ë ˆì´ì§• í™•ë¥  ê²°ê³¼ê°’ ë„ì¶œ

# íŒ¨ëŸ¬í”„ë ˆì´ì¦ˆê°€ ë§ëŠ” ê²½ìš°
for i in range(len(classes)):
    print(f"{classes[i]}: {int(round(paraphrase_results[i] * 100))}%")

# íŒ¨ëŸ¬í”„ë ˆì´ì¦ˆê°€ ì•„ë‹Œ ê²½ìš° 
for i in range(len(classes)):
    print(f"{classes[i]}: {int(round(not_paraphrase_results[i] * 100))}%")
```

## 2. Extractive Question Answering(ì§ˆì˜ ì‘ë‹µ)

> ì§ˆì˜ ì‘ë‹µì€ ì§ˆë¬¸ì´ ì£¼ì–´ì¡Œì„ ë•Œ í…ìŠ¤íŠ¸ë¡œë¶€í„° ë‹µë³€ì„ ì¶”ì¶œí•˜ëŠ” ê³¼ì—…ìœ¼ë¡œì„œ ì§ˆì˜ ì‘ë‹µ ë°ì´í„°ì…‹ì˜ ì˜ˆì‹œë¡œëŠ”
**SQuAD(The Stanford Question Answering Dataset)**ì´ ìˆìœ¼ë©° ì´ëŠ” ì „ì ìœ¼ë¡œ ì´ ê³¼ì—…ì„ ëª©ì ìœ¼ë¡œ í•œ ë°ì´í„°ì…‹ì„. ì´ ë°ì´í„°ì…‹ì„ ë¯¸ì„¸ ì¡°ì •í•˜ê³ ì í•˜ëŠ” ê²½ìš°ì—ëŠ” [run_qa.py](https://github.com/huggingface/transformers/tree/master/examples/pytorch/question-answering/run_qa.py) and
[run_tf_squad.py](https://github.com/huggingface/transformers/tree/master/examples/tensorflow/question-answering/run_tf_squad.py) ìŠ¤í¬ë¦½íŠ¸ë¥¼ í™œìš©í•  í•„ìš”ê°€ ìˆìŒ.

>ì•„ë˜ì˜ ì‚¬ë¡€ëŠ” ì§ˆì˜ ì‘ë‹µì„ í•˜ê¸° ìœ„í•´ì„œ íŒŒì´í”„ë¼ì¸ì„ ì‚¬ìš©í•œ ì‚¬ë¡€ë¡œì„œ ì£¼ì–´ì§„ í…ìŠ¤íŠ¸ì—ì„œ ì§ˆë¬¸ì— ëŒ€í•œ ë‹µë³€ì„ ì¶”ì¶œí•´ë‚´ëŠ” ê²ƒì´ë©°, SQuAD ë°ì´í„°ì…‹ì— ëŒ€í•´ ë¯¸ì„¸ ì¡°ì •ëœ ëª¨ë¸ì„ í™œìš©í•˜ì˜€ìŒ. 

```python
from transformers import pipeline

# ì§ˆì˜ì‘ë‹µì— ëŒ€í•œ íŒŒì´í”„ë¼ì¸ êµ¬ì¶•
question_answerer = pipeline("question-answering")

# í…ìŠ¤íŠ¸ ì œì‹œ 
context = r"""
Extractive Question Answering is the task of extracting an answer from a text given a question. An example of a
question answering dataset is the SQuAD dataset, which is entirely based on that task. If you would like to fine-tune
a model on a SQuAD task, you may leverage the examples/pytorch/question-answering/run_squad.py script.
"""
```

> ì•„ë˜ì—ì„œëŠ” í…ìŠ¤íŠ¸ë¡œë¶€í„° "ì‹œì‘"ê³¼ "ë" ê°’ì„ ë”°ë¼ ë‹µê³¼ ì‹ ë¢°ì„± ì ìˆ˜ë¥¼ ë„ì¶œí•´ë‚´ëŠ”ë° ì´ëŠ” í…ìŠ¤íŠ¸ì— ìˆëŠ” ë‹µì˜ ìœ„ì¹˜ë¥¼ ëœ»í•¨. 

```python
result = question_answerer(question="What is extractive question answering?", context=context)
print(f"Answer: '{result['answer']}', score: {round(result['score'], 4)}, start: {result['start']}, end: {result['end']}")
result = question_answerer(question="What is a good example of a question answering dataset?", context=context)
print(f"Answer: '{result['answer']}', score: {round(result['score'], 4)}, start: {result['start']}, end: {result['end']}")
```

```python
# ëª¨ë“ˆ ë¶ˆëŸ¬ì˜¤ê¸° 
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
import torch

# ëª¨ë¸ê³¼ í† í¬ë‚˜ì´ì € ì„¤ì •í•˜ê¸°
tokenizer = AutoTokenizer.from_pretrained("bert-large-uncased-whole-word-masking-finetuned-squad")
model = AutoModelForQuestionAnswering.from_pretrained("bert-large-uncased-whole-word-masking-finetuned-squad")

# í…ìŠ¤íŠ¸ ì„¤ì •í•˜ê¸° 
text = r"""
ğŸ¤— Transformers (formerly known as pytorch-transformers and pytorch-pretrained-bert) provides general-purpose
architectures (BERT, GPT-2, RoBERTa, XLM, DistilBert, XLNetâ€¦) for Natural Language Understanding (NLU) and Natural
Language Generation (NLG) with over 32+ pretrained models in 100+ languages and deep interoperability between
TensorFlow 2.0 and PyTorch.
"""
# ì§ˆë¬¸ ì œì‹œí•˜ê¸° 
questions = [
    "How many pretrained models are available in ğŸ¤— Transformers?",
    "What does ğŸ¤— Transformers provide?",
    "ğŸ¤— Transformers provides interoperability between which frameworks?",
]
for question in questions:
    inputs = tokenizer(question, text, add_special_tokens=True, return_tensors="pt")
    input_ids = inputs["input_ids"].tolist()[0]

    outputs = model(**inputs)
    answer_start_scores = outputs.start_logits
    answer_end_scores = outputs.end_logits

    answer_start = torch.argmax(
        answer_start_scores
    )  # ìµœê³  ì ìˆ˜ê°’ìœ¼ë¡œì„œ ê°€ì¥ ëŒ€ë‹µì˜ ì‹œì‘ì´ ë  ë¶€ë¶„ì„ ì–»ìŒ  
    answer_end = torch.argmax(answer_end_scores) + 1  # ìµœê³  ì ìˆ˜ê°’ìœ¼ë¡œì„œ ê°€ì¥ ëŒ€ë‹µì˜ ëì´ ë  ë¶€ë¶„ì„ ì–»ìŒ  

    answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end]))

    # ì§ˆë¬¸ê³¼ ë‹µì„ ì¶œë ¥í•˜ê¸° 
    print(f"Question: {question}")
    print(f"Answer: {answer}")
```

> **ì•„ë˜ëŠ” ëª¨ë¸ê³¼ í† í¬ë‚˜ì´ì €ë¥¼ í™œìš©í•´ì„œ ë§ˆìŠ¤í‚¹ ì–¸ì–´ ëª¨ë¸ë§ì„ í•˜ëŠ” ì‚¬ë¡€ë¡œì„œ ê³¼ì •ì€ ë‹¤ìŒê³¼ ê°™ìŒ.**

1. ì²´í¬í¬ì¸íŠ¸ë¡œë¶€í„° í† í¬ë‚˜ì´ì €ì™€ ëª¨ë¸ì„ ê°€ë™í•˜ëŠ”ë° ì´ ë•Œ ëª¨ë¸ì€ DistilBERTì´ë©° ì´ë¥¼ ì²´í¬í¬ì¸íŠ¸ì— ì €ì¥ëœ ê°€ì¤‘ì¹˜ê°’ê³¼ í•¨ê»˜ ì˜¬ë¦¼
2. ë§ˆìŠ¤í‚¹ëœ í† í°ê³¼ í•¨ê»˜ ì‹œí€€ìŠ¤ë¥¼ ì •ì˜í•˜ë©°, ë‹¨ì–´ ëŒ€ì‹ ì— `tokenizer.mask_token`ë¥¼ ë°°ì¹˜ì‹œí‚´
3. IDì˜ ë¦¬ìŠ¤íŠ¸ì— ì‹œí€€ìŠ¤ë¥¼ ì¸ì½”ë”©í•˜ë©°, í•´ë‹¹ ë¦¬ìŠ¤íŠ¸ì—ì„œ ë§ˆìŠ¤í‚¹ëœ í† í°ì˜ ìœ„ì¹˜ë¥¼ ì°¾ì•„ëƒ„
4. ë§ˆìŠ¤í‚¹ í† í°ì˜ ì¸ë±ìŠ¤ì—ì„œ ì˜ˆì¸¡ê°’ì„ êµ¬í•  ë•Œ, ì´ í…ì„œê°’ì€ ê·¸ ë‹¨ì–´ì™€ ë™ì¼í•œ í¬ê¸°ì—¬ì•¼ í•¨. ë˜í•œ ê·¸ ê°’ì€ ê°ê°ì˜ í† í°ì— ì†í•˜ëŠ” ì ìˆ˜ë¡œì„œ ê·¸ ë¬¸ë§¥ì—ì„œ ë³´ë‹¤ ê·¸ëŸ´ ë“¯í•œ í† í°ì— ëª¨ë¸ì´ ì ìˆ˜ë¥¼ ë¶€ì—¬í•¨
5. íŒŒì´í† ì¹˜ì˜ `topk`ë‚˜ í…ì„œí”Œë¡œìš°ì˜ `top_k` ë©”ì„œë“œë¥¼ í™œìš©í•˜ì—¬ ìƒìœ„ 5ê°œì˜ í† í°ì„ ì–»ì–´ëƒ„.
6. ê·¸ (5ê°œì˜ í† í°ë“¤ë¡œ) ë§ˆìŠ¤í‚¹ í† í°ì„ ëŒ€ì²´í•˜ê³  ê²°ê³¼ë¥¼ ì¶œë ¥í•¨

```python
from transformers import AutoModelWithLMHead, AutoTokenizer
import torch
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-cased")
model = AutoModelWithLMHead.from_pretrained("distilbert-base-cased")
sequence = f"Distilled models are smaller than the models they mimic. Using them instead of the large versions would help {tokenizer.mask_token} our carbon footprint."
input = tokenizer.encode(sequence, return_tensors="pt")
mask_token_index = torch.where(input == tokenizer.mask_token_id)[1]
token_logits = model(input).logits
mask_token_logits = token_logits[0, mask_token_index, :]
top_5_tokens = torch.topk(mask_token_logits, 5, dim=1).indices[0].tolist()
```

> ì´ëŠ” ê²°ê³¼ê°’ìœ¼ë¡œ ëª¨ë¸ì— ì˜í•´ì„œ ì˜ˆì¸¡ëœ 5ê°œ í† í°ê³¼ í•¨ê»˜ ë‹¤ì„¯ ê°œì˜ ë¬¸ìì—´ì„ ì¶œë ¥í•¨.

```python
for token in top_5_tokens:
    print(sequence.replace(tokenizer.mask_token, tokenizer.decode([token])))
```

### Causal Language Modeling (ì¸ê³¼ê´€ê³„ ì–¸ì–´ëª¨ë¸)

> ì¸ê³¼ê´€ê³„ ì–¸ì–´ëª¨ë¸ì€ íŠ¹ì • í† í° ë’¤ì— ì´ì–´ì§€ëŠ” í† í°ë“¤ì„ ì˜ˆì¸¡í•˜ëŠ” ê³¼ì—…ì„. ì´ ë•Œ, ëª¨ë¸ì€ ì˜¤ë¡œì§€ (ë§¤ìŠ¤í‚¹ëœ í† í°ì˜ ì™¼ìª½ í† í°ì¸) ë¬¸ë§¥ì˜ ì™¼ìª½ ë§Œì„ ë‹¤ë£¨ê²Œ ë¨. ì´ëŸ¬í•œ í›ˆë ¨ê³¼ì •ì€ ìƒì„± ê³¼ì—…ì— ìˆì–´ì„œëŠ” ìƒë‹¹íˆ í¥ë¯¸ë¡œìš´ ì¼ì´ë©° ì¸ê³¼ê´€ê³„ ì–¸ì–´ëª¨ë¸ì— ëŒ€í•´ì„œ ë¯¸ì„¸ì¡°ì •ì„ í•˜ê³  ì‹¶ë‹¤ë©´, :prefix_link:*run_clm.py <examples/pytorch/language-modeling/run_clm.py>* ì„ í™œìš©í•  ìˆ˜ ìˆìŒ.

> í”íˆ, ë°”ë¡œ ë’¤ì— ë‚˜ì˜¤ëŠ” í† í°ì€ ì…ë ¥ ì‹œí€€ìŠ¤ì—ì„œ ëª¨ë¸ì´ ìƒì„±í•œ ë§ˆì§€ë§‰ ì€ë‹‰ ìƒíƒœì˜ ë¡œì§“ê°’ìœ¼ë¡œë¶€í„° ìƒ˜í”Œë§ì´ ë˜ì–´ ì˜ˆì¸¡ë¨. 

> ëª¨ë¸ê³¼ í† í¬ë‚˜ì´ì €ë¥¼ í™œìš©í•˜ê³  í† í°ì˜ ì…ë ¥ ì‹œí€€ìŠ¤ë¥¼ ì‡ëŠ” ê·¸ ë‹¤ìŒ í† í°ì„ ìƒ˜í”Œë§í•˜ëŠ” ë°©ë²•ìœ¼ë¡œ `PreTrainedModel.top_k_top_p_filtering`ì„ í™œìš©í•˜ëŠ” ì‚¬ë¡€ê°€ ì•„ë˜ì— ë‚˜ì™€ ìˆìŒ.

```python
from transformers import AutoModelWithLMHead, AutoTokenizer, top_k_top_p_filtering
import torch
from torch import nn

# í† í¬ë‚˜ì´ì €ì™€ ëª¨ë¸ ì„¤ì •í•˜ê¸°
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelWithLMHead.from_pretrained("gpt2")

# ì‹œí€€ìŠ¤ ì œì‹œí•˜ê¸° 
sequence = f"Hugging Face is based in DUMBO, New York City, and"

# ì…ë ¥ê°’ ì¸ì½”ë”©í•˜ê¸° 
input_ids = tokenizer.encode(sequence, return_tensors="pt")

# ë§ˆì§€ë§‰ ì€ë‹‰ìƒíƒœë¡œë¶€í„° ë¡œì§“ê°’ ì–»ê¸°
next_token_logits = model(input_ids).logits[:, -1, :]

# í•„í„°ë§í•˜ê¸°
filtered_next_token_logits = top_k_top_p_filtering(next_token_logits, top_k=50, top_p=1.0)

# í™•ë¥ ê°’ êµ¬í•˜ê¸°
probs = nn.functional.softmax(filtered_next_token_logits, dim=-1)

# ë‹¤ìŒ í† í° êµ¬í•˜ê¸° 
next_token = torch.multinomial(probs, num_samples=1)
generated = torch.cat([input_ids, next_token], dim=-1)

# ê²°ê³¼ê°’ êµ¬í•˜ê¸° 
resulting_string = tokenizer.decode(generated.tolist()[0])
```

> ì—¬ê¸°ì„œëŠ” ê¸°ì¡´ ì‹œí€€ìŠ¤ë¥¼ ì‡ëŠ” ë‹¤ìŒ ì½”íŠ¼ì„ ë‚˜ì˜¤ë©°, ì´ ì‚¬ë¡€ì—ì„œëŠ” 'and' ë’¤ì˜ í† í°ì„ ì˜ë¯¸í•¨.

```python
print(resulting_string)
```
